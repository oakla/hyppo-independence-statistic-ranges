```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import numpy as np

def multi_normal_viz(size1, size2, dpi,mu=np.array([1,2]),cov=np.array([[.5, .25],[.25, .5]]), range=2):
    x = np.linspace(mu[0]-range, mu[0]+range, 100)
    y = np.linspace(mu[1]-range, mu[1]+range, 100)
    X, Y = np.meshgrid(x, y)
    pos = np.dstack((X, Y))

    rv = multivariate_normal(mu, cov)
    Z = rv.pdf(pos)
    fig = plt.figure(figsize=(size1,size2))
    # edits
    fig.set_dpi(dpi)
    # end edits
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z)
    fig.show()
```

```python
multi_normal_viz(10,10,100)
```

```python
multi_normal_viz(4,4,200)
```

```python
multi_normal_viz(1,1,200)
```

```python
multi_normal_viz(2,2,200)
```
