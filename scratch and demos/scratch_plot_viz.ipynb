```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import numpy as np
```

```python
def multi_normal_viz(mu=np.array([1,2]),cov=np.array([[.5, .25],[.25, .5]]), range=3):
    x = np.linspace(mu[0]-range, mu[0]+range, 50)
    y = np.linspace(mu[1]-range, mu[1]+range, 50)
    X, Y = np.meshgrid(x, y)
    pos = np.dstack((X, Y))

    rv = multivariate_normal(mu, cov)
    Z = rv.pdf(pos)
    fig = plt.figure(figsize=(10,10))
    # edits
    fig.set_dpi(100)
    # end edits
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z)
    fig.show()
```

```python
multi_normal_viz()
```

```python
mu=np.array([0,1])
cov=np.array([[10, 5],[2, 2]])
multi_normal_viz(mu, cov,10)
```

```python
mu=np.array([0,1])
cov=np.array([[10, -5],[-2, 2]])
multi_normal_viz(mu, cov,10)
```
